## Sombrero

Below is the sombrero plot.  We can imagine the system we are modelling as a ball somewhere in the sombrero.  With drag, we would expect that the ball will eventually rest in one of the minima, but things get interesting when we add a term that is essentially moving the hat from left to right.  In order to study this system we solved the following pair of ODEs using the 4th order Runge-Kutta method:

$$\dot{x}(t) = y(t)$$
$$m\dot{y}(t) = -\nu y(t) + x(t) - x^3(t) + F\cos(\omega t)$$

This Runge-Kutta method was coded in "sombrero.py" and numba.jit was applied to significantly speed up the Runge-Kutta method.

```python
som.plot_sombrero()
```

## 1.
It appears that for the set of initial conditions and the driving force factor .18, that the position of the ball seems to reach a steady state.  The parametric curve shows that there are about 10 equilibrium points that the ball "prefers" to sit around. It is also important to note that the ball seems to just stay on the left side and isn't going over the middle "hill" in the sombrero (or potential energy).

```python
som.plot_x(-.9,0,.18,314159)
```

```python
som.plot_parametric(-.9,0,.18,50)
```

## 2.

When we change the value of x(0), it appears that the ball hovers between .8 and 1.2 in terms of side to side movement on the x-axis rather than between -1.2 and -.8, which makes sense because we started the ball on the right side of the sombrero, whereas in the first case we set it on the left side.  It appears that the ball likes to settle in whichever side we place it in first.

```python
som.plot_x(.9,0,.18,314159)
```

```python
som.plot_parametric(.9,0,.18,50)
```

## 3.

When we change the initial conditions and increase the force constant it appears that the ball changes position on the x-axis much more dramatically at the beginning, before settling into a pretty small area of width about .8, like we saw before.  So while it takes longer to settle and it looks like it changes position more dramatically at the start, the steady state behavior is relatively similar to how it was in problems 1 and 2.  One interesting thing to note is that a change of .001 in x(0) causes the ball to settle in a space that is completely different from with x(0) = .2. In the first case the ball is settling on the right side of the sombrero and in the second case, the ball is settling in the left side of the sombrero.  The poincare section also shows that there are more steady state points (ball oscillates between more points).
